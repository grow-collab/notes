```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.rcParams['font.sans-serif'] = ['FangSong']

# 第一种方式
x = list(range(7))
y1 = [10,9,8,7,5,4,3]
y2 = [9,5,8,3,6,4,8]

fig,ax = plt.subplots() # 创建画布和区域
ax.bar(x,y1,width=0.4,edgecolor='black',label='系列1')
ax.bar([i+0.4 for i in x],y2,width=0.4,edgecolor='black',label='系列2')
ax.legend(loc=1)
ax.set_xlabel('x轴')
ax.set_ylabel('y轴')
plt.xticks(x,['周一','周二','周三','周四','周五','周六','周日'])
ax.set_title('分组柱状图一')
plt.show()
```

```python
# 第二种方式
x = np.array(list(range(7)))
y1 = [7,6,5,4,3,2,1]
y2 = [9,8,7,6,5,4,3]
plt.figure(figsize=(5,5))
plt.bar(x,y1,width=0.4,edgecolor='black',label='系列一')
plt.bar(x+0.4,y2,width=0.4,edgecolor='black',label='系列二')
plt.legend(loc=1)
plt.xlabel('x轴')
plt.ylabel('y轴')
plt.title('分组柱状图二')
plt.xticks(x,['周一','周二','周三','周四','周五','周六','周日'])
plt.show()
```
