```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
fig,ax = plt.subplots()

# marker,改变点位的形状
# alpha,调整线的透明度
ax.plot([1,2,3],[4,5,6],marker='v',alpha=0.5) # *,d,v等
plt.show()
```

```python
x = np.arange(1,10,0.1)
y1 = np.sin(x)
y2 = np.cos(x)
fig,ax = plt.subplots() # 区域和画布

# linewidth,设置线条宽度
#  alpha,设置透明度
# markersize,点位形状的大小
ax.plot(x,y1,alpha=0.7,marker='*',linewidth=10,markersize=5)
ax.plot(x,y2,alpha=0.2)
plt.show()
```
