```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']

x = ['周一','周二','周三','周四','周五','周六','周日']
y1 = [7,6,5,4,3,2,1]
y2 = [1,2,3,3,1,1,0.5]
fig,ax = plt.subplots() # 创建画布和区域
ax.bar(x=x,height=y1,edgecolor='red',label='系列1') # edgecolor:边框颜色
ax.bar(x=x,height=y2,edgecolor='green',label='系列2')
ax.legend(loc=1,fontsize=20) # fontsize 设置大小
ax.set_ylabel('y轴',fontsize=16)
ax.set_xlabel('x轴',fontsize=16)
ax.set_title('堆叠柱状图')

for i,j,q in zip(x,y1,y2):
    print(i,j,q)
    p1 = j/(q+j)
    p2 = q/(q+j) # y2的柱子所占整个柱子的比例
    
    plt.text(i,j*0.4,'{:.0%}'.format(p2),ha='center',fontsize=16)
    plt.text(i,j,'{:.0%}'.format(p1),ha='center',fontsize=16)
plt.show()
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']

x = ['周一','周二','周三','周四','周五','周六','周日']
y1 = [7,6,5,4,3,2,1]
y2 = [1,2,3,3,1,1,0.5]
fig,ax = plt.subplots() # 创建画布和区域
ax.bar(x=x,height=y1,edgecolor='red',label='系列1') # edgecolor:边框颜色
ax.bar(x=x,height=y2,edgecolor='green',label='系列2',bottom=y1)
ax.legend(loc=1,fontsize=20) # fontsize 设置大小
ax.set_ylabel('y轴',fontsize=16)
ax.set_xlabel('x轴',fontsize=16)
ax.set_title('堆叠柱状图')

for i,j,q in zip(x,y1,y2):
    print(i,j,q)
    p1 = j/(q+j)
    p2 = q/(q+j) # y2的柱子所占整个柱子的比例
    
    plt.text(i,j*0.4,'{:.0%}'.format(p1),ha='center',fontsize=16)
    plt.text(i,j+q*0.4,'{:.0%}'.format(p2),ha='center',fontsize=16)
plt.show()
```
