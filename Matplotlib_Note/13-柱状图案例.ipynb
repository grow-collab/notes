```python
import matplotlib.pyplot as plt
```

```python
# 设置中文
plt.rcParams['font.sans-serif'] = ['SimHei']

x = list(range(7))
y = [127, 127, 176, 200, 154, 78, 159]

fig, ax = plt.subplots()  # 创建画布和区域
ax.bar(x, y, width=0.4, edgecolor='black', alpha=0.7)

yticks = list(range(0, 350, 50))  # 设置y轴刻度
plt.yticks(yticks)

xticks = ['小蓝', '小绿', '小黄', '小红', '小陈', '小龙', '小灰']
plt.xticks(x, xticks,rotation=60) # rotation 样式旋转为60°

ax.set_ylabel('分数')
ax.set_title('成绩总和',fontsize=16)
for i, j in zip(x, y):
    plt.text(i, j + 4, j, ha='center')
plt.show()
```
