```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']

value = list(range(10))
ks = list('qwertyuiop')
fig = plt.figure(figsize=(8, 4))  # figsize = 设置宽和高
ax = fig.add_subplot(121)  # 1行 2列 在第一个
ax.set_title('图1 垂直图')
ax.bar(x=value,
       height=ks,
       alpha=0.5,
       width=0.5,
       color='g',
       )  # bar 垂直图 x 和 height
"""
alpha:透明度
width:线条宽度
color:颜色
edgecolor:边框颜色
linewidth:边框宽度
"""

ax2 = fig.add_subplot(122)
ax2.set_title('图2 水平图')
ax2.barh(y=ks, 
         width=value,
         alpha=0.5,
         color=['green','blue','yellow','pink','gray','magenta'])  # barh 水平图 y 和 width
plt.show()
```
