```python
fig = plt.figure() # 创建画布
ax = fig.add_subplot() # 在画布添加区域
x = ['上海','北京','武汉']
y = np.array([1,2,3])

# linestyle,默认实线(-)
# -- 虚线, -. 长短点线, : 点线,
ax.plot(x,y+3,color="r",linestyle="--") # 可通过十六进制值,英文单词或其首字母
plt.show()
```
