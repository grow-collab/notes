```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
fig,ax = plt.subplots()
ax.plot([0,1,3],[1,3,4],'-.k')
ax.plot([0,1,3],[3,5,6],'-g',marker='v')
plt.xlim(0,3) # 设置x轴坐标范围
plt.ylim(3,9) # 设置x轴坐标范围
plt.show()
```

```python
fig,ax = plt.subplots()
ax.plot([0,1,3],[1,3,4],'-.k')
ax.plot([0,1,3],[3,5,6],'-g',marker='v')
plt.axis([0,3,1,9]) # [xmin,xmax,ymin,ymax]
plt.show()
```
