```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()  # 创建画布和区域

# 准备数据
x = list(range(1, 8))  # x轴的数值,列表的形式 周

# 两个城市的气温数据
y_wuhan = [15, 20, 22, 23, 20, 18, 16]
y_heilongjiang = [-10, -8, -12, -10, -8, -6, 1]

plt.rcParams['font.sans-serif'] = ['FangSong'] # 宋体
plt.rcParams['axes.unicode_minus'] = False # 特殊字符也会显示

ax.plot(x, y_wuhan, color="r", linestyle=":", linewidth=5, marker="*", markersize=10,alpha=0.8,label="武汉") # label:线标签
ax.plot(x, y_heilongjiang, color="g", linestyle="-.", linewidth=3, marker="d", markersize=10,label="黑龙江")

ax.set_title('武汉和黑龙江的天气走势')
plt.xlabel('日期') # 设置x轴名字
plt.ylabel('温度') # 设置y轴名字
plt.legend(loc=1)

# 创建修改的x的值
xticks = ['周一','周二','周三','周四','周五','周六','周日']
plt.xticks(x,xticks,rotation=45) # x是原先的值,xticks是修改后的值,rotation 旋转的角度

yticks = list(range(-15,35,5))
plt.yticks(yticks)

# 显示度数
# text
for i,j in zip(x,y_wuhan):
    # print(i,j)
    plt.text(i,j+2,f'{j}℃',horizontalalignment='center')
    
for i,j in zip(x,y_heilongjiang):
    # print(i,j)
    plt.text(i,j+2,f'{j}℃',horizontalalignment='center')
plt.show()
```
