```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.rcParams['font.sans-serif'] = ['FangSong']
# SimHei 微软雅黑
# FangSong 宋体

fig,ax = plt.subplots() # 创建区域和画布
ax.plot(['上海','武汉','北京','邯郸','重庆','广东'],[1,2,3,4,5,45]) # 在区域里作图
# 设置标题
ax.set_title('I Love You',fontdict={'size':16},loc='right') # left,左  center,居中  right,右
# 设置轴标签
ax.set_xlabel('x轴标签')
ax.set_ylabel('y轴标签')

# 设置刻度
ax.set_xticks([0,1,2,3,4,5])
ax.set_yticks([0,10,20,30,40,50])
plt.show()
```

```python
fig,ax = plt.subplots()
# 第一种 设置图例 label 和legend 必须要一起使用
ax.plot(['上海','北京','武汉'],[1,3,5],label='2023')
ax.plot(['上海','北京','武汉'],[2,4,6],label='2024')
ax.set_title('规划',fontdict={'size':16},loc='right')
ax.legend()
plt.show()
```

```python
# 显示网格
fig,ax = plt.subplots()
ax.plot([1,2,3],[4,5,6],color='r')
ax.grid(axis='y') # 参数指定轴,什么都不写的情况下,x,y都会显示网格线
# ax.grid()
plt.show()
```
